```python
import numpy as np
import pandas as pd
import random
from sklearn.model_selection import train_test_split
import xgboost
```

```python

companies = [
    "Apple", "Oppo", "Vivo", "Xiaomi", "OnePlus", "Poco", "Samsung", "Nokia"
]
company_number = [1,2,3,4,5,6,7,8]

ram_sizes = [4, 6, 8]
storage = [64, 128, 256]
battery = np.arange(3000, 8001, 100)

data = {
    "Company": [random.choice(company_number) for _ in range(10000)], 
    "RAM": [random.choice(ram_sizes) for _ in range(10000)],
    "Storage": [random.choice(storage) for _ in range(10000)],
    "Battery": [random.choice(battery) for _ in range(10000)],
}


df = pd.DataFrame(data)

ram_weight = 50
storage_weight = 30
battery_weight = 0.01

company_base_prices = {
    1: 15000, #Apple
    2: 8000, #Samsumg
    3: 6000, #Oneplus
    4: 4000, #xiaomi
    5: 3500, #oppo
    6: 3000, #vivo
    7: 2500, #poco
    8: 2000 #nokia
}


def estimate_price(row):
    base_price = company_base_prices.get(row["Company"], 200) 
    ram_price = row["RAM"] * ram_weight *3
    storage_price = row["Storage"] * storage_weight*2
    battery_price = (row["Battery"] - 3000) * battery_weight*2  
    return base_price + ram_price + storage_price + battery_price

df["Price"] = df.apply(estimate_price, axis=1)
print(df.head())

df.to_csv("dataset.csv", index=False)
```
